# $$\textbf{Teori}$$

### Bildiğimiz üzere elektirk alan çizgilerini oluşturacak formül:
### $$\vec E = \frac{1}{4\pi\epsilon_{0}} \frac{q}{r^2}\hat r $$
### Bir noktanın başka bir noktaya olan en kısa uzaklığı şu şekilde ifade edilebilir:
### $$ r = \sqrt{x^2 + y^2}$$
### lineer cebirden biliyoruz ki bir yön vektörü şu şekilde açılabilir:
### $$\hat r = \frac{\vec r}{|\vec r|} $$
### $$ E = \frac{1}{4\pi\epsilon_{0}} \frac{q}{r^3}\vec r  $$
### buradaki payda kısmı:
### $$ r^3 = (\sqrt{x^2 + y^2})^3 = (x^2 + y^2)^{3/2} $$
### burada x dedğimiz şey parçağımızın yatay eksendeki konumu eksi etki edeceği noktanın yatay eksendeki konumunun mutlak değeridir, aynısı y içinde geçerlidir.



```python
import matplotlib.pyplot as plt
import numpy as np
import random

#Bir elektrik alan çizme projesi


def Elektrik_alan_ciz(yuk,N,M):
    #koordinatlar 
    X = np.arange(0, M, 1)
    Y = np.arange(0, N, 1)
    X, Y = np.meshgrid(X, Y)

    konum_x = float(input('x konumu girin'))
    konum_y = float(input('y konumu girin'))

    Ex ,Ey = np.zeros((N,M)), np.zeros((N,M)) #Elektrik alanımın x ve y si için depo 

    konum = [[], []] 

    konum[0].append(konum_x)
    konum[1].append(konum_y)

    for i in range(N):
        for j in range(M):
            payda = ((i - konum_y)**2 + (j - konum_x)**2)**1.5  #payda = denominator, denom of E. field equation is ((x-x_0)^2 + (y - y_0)^2)*\sqrt((x-x_0)^2 + (y - y_0)^2)
            if payda != 0:
                Ex[i, j] += yuk*(8.99*10**9)*(j - konum_x) / payda # k = 8.99*10**9
                Ey[i, j] += yuk*(8.99*10**9)*(i - konum_y) / payda

    
    fig, ax = plt.subplots(figsize = (12,8))
    ax.add_artist(plt.Circle((konum_x,konum_y), 0.17, color='r'))
    q = ax.streamplot(X, Y, Ex, Ey)
    #plt.figure(figsize = (12, 8))
    #plt.plot(*konum, 'bo')
    #plt.streamplot(X, Y, Ex, Ey)
    plt.title('Elektirk Alan')
    #plt.axis('equal')
    plt.show()
```

```python
Elektrik_alan_ciz(25*10**(-9),25,25)

Elektrik_alan_ciz(-25*10**(-9),25,25)
```
