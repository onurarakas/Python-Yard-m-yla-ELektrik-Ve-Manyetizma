```python
import matplotlib.pyplot as plt
import numpy as np
import random

plt.style.use('dark_background')

#Bir elektrik alan çizme projesi
a = np.linspace(10,20,100000)
b = np.ones(100000)
konumlar = np.empty(len(a), dtype=object)
for i in range(len(a)):
    konumlar[i] = (a[i], 20*b[i])

def Elektrik_alan_ciz(konumlar,yuk,M,N):
    yuk = float(yuk)


    if yuk > 0.0:
        point_color = 'r'
    else:
        point_color = 'b'
    #koordinatlar 
    X = np.arange(0, M, 1)
    Y = np.arange(0, N, 1)
    X, Y = np.meshgrid(X, Y)

    Ex = np.zeros_like(X, dtype=np.float64)
    Ey = np.zeros_like(Y, dtype=np.float64)  #Elektrik alanımın x ve y si için depo 

    # ilk kod da 

    payda = []

    for konum in konumlar:    
        for i in range(N):
            for j in range(M):
                payda = ((i - konum[1])**2 + (j - konum[0])**2)**1.5  #payda = denominator, denom of E. field equation is ((x-x_0)^2 + (y - y_0)^2)*\sqrt((x-x_0)^2 + (y - y_0)^2)
                if payda != 0:  # Check if payda is not zero to avoid division by zero   
                    Ex[i, j] += yuk*(8.99*10**9)*(j - konum[0]) / payda # k = 8.99*10**9
                    Ey[i, j] += yuk*(8.99*10**9)*(i - konum[1]) / payda

        
    fig, ax = plt.subplots(figsize = (12,8))
    for konum in konumlar:
        ax.add_artist(plt.Circle((konum[0],konum[1]), 0.17, color=point_color))
    q = ax.streamplot(X, Y, Ex, Ey, density=4, linewidth = 1.5, arrowsize=0.5)
    #plt.figure(figsize = (12, 8))
    #plt.plot(*konum, 'bo')
    #plt.streamplot(X, Y, Ex, Ey)
    plt.title('Elektirk Alan')
    #plt.axis('equal')
    plt.show()
```

```python
Elektrik_alan_ciz(konumlar,1,40,40)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an empty array for kare_konumlar
kare_konumlar = np.empty(4000, dtype=object)

# Assign values to the slices to create a square
# First side
kare_konumlar[0:1000] = [(x, 40.0) for x in np.linspace(10.0, 40.0, 1000)]
# Second side
kare_konumlar[1000:2000] = [(10.0, y) for y in np.linspace(40.0, 70.0, 1000)]
# Third side
kare_konumlar[2000:3000] = [(x, 70.0) for x in np.linspace(10.0, 40.0, 1000)]
# Fourth side
kare_konumlar[3000:4000] = [(40.0, y) for y in np.linspace(70.0, 40.0, 1000)]

# Extract x and y coordinates
x_coords = [point[0] for point in kare_konumlar if point is not None]
y_coords = [point[1] for point in kare_konumlar if point is not None]

# Plot the points
plt.plot(x_coords, y_coords, 'ro')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Square with specified coordinates')
plt.gca().set_aspect('equal', adjustable='box')  
plt.show()
# This code is writen by chatus gptus
```
