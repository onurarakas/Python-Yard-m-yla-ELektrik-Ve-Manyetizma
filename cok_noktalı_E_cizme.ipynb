```python
import matplotlib.pyplot as plt
import numpy as np
import random

#Bir elektrik alan çizme projesi

a = np.linspace(25,40,100)
b = np.ones(100)
konumlar = np.empty(len(a), dtype=object)
for i in range(len(a)):
    konumlar[i] = (a[i], 20*b[i])

def Elektrik_alan_ciz(konum,yuk,N,M):
    #koordinatlar 
    X = np.arange(0, M, 1)
    Y = np.arange(0, N, 1)
    X, Y = np.meshgrid(X, Y)

    shape = konum.shape[0]
    Ex ,Ey = np.zeros((shape,shape)), np.zeros((shape,shape)) #Elektrik alanımın x ve y si için depo 

    payda = []

    for konum in konumlar:    
        for i in range(N):
            for j in range(M):
                payda = ((i - konum[1])**2 + (j - konum[0])**2)**1.5  #payda = denominator, denom of E. field equation is ((x-x_0)^2 + (y - y_0)^2)*\sqrt((x-x_0)^2 + (y - y_0)^2)
                    
                Ex[i, j] += yuk*(8.99*10**9)*(j - konum[0]) / payda # k = 8.99*10**9
                Ey[i, j] += yuk*(8.99*10**9)*(i - konum[1]) / payda

        
    fig, ax = plt.subplots(figsize = (12,8))
    for konum in konumlar:
        ax.add_artist(plt.Circle((konum[0],konum[1]), 0.17, color='r'))
    q = ax.streamplot(X, Y, Ex, Ey, density=3)
    #plt.figure(figsize = (12, 8))
    #plt.plot(*konum, 'bo')
    #plt.streamplot(X, Y, Ex, Ey)
    plt.title('Elektirk Alan')
    #plt.axis('equal')
    plt.show()
```

```python
Elektrik_alan_ciz(konum,1,100,100)
```
